```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
%matplotlib inline
```

```python
df=pd.read_csv('Bangalore_Restaurants.csv')
df.dropna()
```

```python
df.isnull().any()
```

```python
df.rename(columns = {'Dining_Rating':'DiningRating'}, inplace = True)
df.rename(columns = {'Dining_Review_Count':'DiningRC'}, inplace = True)
df.rename(columns = {'Delivery_Rating':'DeliveryRating'}, inplace = True)
df.rename(columns = {'Delivery_Rating_Count':'DeliveryRC'}, inplace = True)
df
```

```python
New_df=df.drop(['Website','Address','Latitude','Longitude','Phone_No'],axis=1)
```

```python
New_df.isnull().sum()
```

```python
A=New_df['DiningRating'].mean()
New_df['DiningRating'].fillna(value=A,inplace=True)
B=New_df['DiningRC'].mean()
New_df['DiningRC'].fillna(value=B,inplace=True)
C=New_df['DeliveryRC'].mean()
New_df['DeliveryRC'].fillna(value=C,inplace=True)
D=New_df['DeliveryRating'].mean()
New_df['DeliveryRating'].fillna(value=D,inplace=True)
```

```python
New_df.isnull().sum()
```

```python
New_df.head(7)
```

```python
srt1 = New_df["Locality"].str.split(",",expand = True)
New_df["Nameofplace"]= srt1[0]
New_df["Area"]= srt1[1]
New_df["city"]=srt1[2]
New_df.drop(columns =["Locality"], inplace = True)
New_df.head(7)
```

```python
New_df['city'].fillna(value='Bangalore',inplace=True)
found = (New_df[New_df['city'].str.contains('Bangalore')])
found
```

```python
(New_df['city'].notnull()).head(7)
```

```python
A = New_df[~New_df["Area"].str.contains('Bangalore')]
A['Area'] = A['Nameofplace'].str.cat(A['Area'],sep=",")
A=A.drop(['Nameofplace','city'],axis=1)
srt2= A["Area"].str.split(",",expand = True)
A["Area"]= srt2[1]
A

```

```python
New_df=New_df.drop(['Area','city'],axis=1)
New_df.rename(columns = {'Nameofplace':'Area'}, inplace = True)
New_df
```

```python
New_df[New_df["Area"].str.contains('BTM')]
```

```python
plt.figure(figsize=(40,50))
sns.countplot(x='Area', data=New_df)
plt.title('Count of Restaurants at each Location', fontsize=40, fontweight='bold')
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(20,40))
df_Restaurant = New_df.value_counts()[:10]
sns.barplot(x= A['Restaurant_Name'], y=A['DiningRC'], palette='Set1')
plt.xlabel("Restaurant_Name", size=15)
plt.ylabel("Dining_Count", size=15)
plt.title("Most famous restaurant chains", fontsize=20, fontweight='bold')
plt.xticks(rotation=90)
plt.show()
```

```python
New_df['Number_of_cuisines_offered'] = New_df['Category'].apply(lambda x : len(x.split(',')))
New_df
```

```python
a = New_df.groupby('Number_of_cuisines_offered').agg({'Pricing_for_2':'mean'})
plt.rcParams["figure.figsize"] = (12,8)
a.plot(kind='bar', color='#7952B3')
plt.title('Approx cost based on Number of cuisines', fontsize=20, fontweight='bold')
plt.ylabel('approx mean cost')
plt.legend()
plt.show()
```
